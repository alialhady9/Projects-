```python
import pandas as pd
import datetime
```

```python
data_1 = pd.read_csv("e-shop clothing 2008.csv",sep = ';')
data_1.head()
```

```python
cols=["year","month","day"]
data_1['date'] = data_1[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis="columns")
data_1.head()
```

```python
data_1['date'] = pd.to_datetime(data_1['date'])
data_1.head()
```

```python
data_1['Weekday'] = data_1['date'].dt.day_name()
data_1.head()
```

```python
country_dict = {1:'Australia',
2:'Austria',
3:'Belgium',
4:'British Virgin Islands',
5:'Cayman Islands',
6:'Christmas Island',
7:'Croatia',
8:'Cyprus',
9:'Czech Republic',
10:'Denmark',
11:'Estonia',
12:'unidentified',
13:'Faroe Islands',
14:'Finland',
15:'France',
16:'Germany',
17:'Greece',
18:'Hungary',
19:'Iceland',
20:'India',
21:'Ireland',
22:'Italy',
23:'Latvia',
24:'Lithuania',
25:'Luxembourg',
26:'Mexico',
27:'Netherlands',
28:'Norway',
29:'Poland',
30:'Portugal',
31:'Romania',
32:'Russia',
33:'San Marino',
34:'Slovakia',
35:'Slovenia',
36:'Spain',
37:'Sweden',
38:'Switzerland',
39:'Ukraine',
40:'United Arab Emirates',
41:'United Kingdom',
42:'USA',
43:'biz', 
44:'com',
45:'int',
46:'net',
47:'org'}
```

```python
data_1.replace({'country': country_dict},inplace = True)
data_1.head()
```

```python
category_dict = {1:'trousers',2:'skirts',3:'blouses',4:'sale'}
colour_dict = {1:'beige',
2:'black',
3:'blue',
4:'brown',
5:'burgundy',
6:'gray',
7:'green',
8:'navy blue',
9:'of many colors',
10:'olive',
11:'pink',
12:'red',
13:'violet',
14:'white'}
location_dict = {1:'top left',2:'top in the middle',3:'top right',4:'bottom left',5:'bottom in the middle',6:'bottom right'}
modelphotography_dict = {1:'en face',2:'profile'}
price2_dict = {1:'yes',2:'no'}
data_1.replace({'page 1 (main category)':category_dict,'colour':colour_dict,'location':location_dict,
               'model photography':modelphotography_dict,'price 2':price2_dict},inplace = True)
```

```python
data_1.head()
```

```python
data_2 = data_1.reindex(columns = ['date','year','month','day','Weekday','order','country','session ID',
                                   'page 1 (main category)','page 2 (clothing model)','colour','location','model photography',
                                   'price','price 2','page'])
data_2.head()
```

```python
data_2.drop(columns = ['order','page 2 (clothing model)','page'],inplace = True)
data_2.head()
```

```python
data_2.rename(columns = {'Weekday':'weekday','page 1 (main category)':'category','price 2':'greaterThanAvgPriceOfCategory'},inplace = True)
data_2.head()
```

```python
data_2.to_csv('CleanedData.csv')
```
